```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)(dont worry about this-done)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Analysis functions

total_schools = len(school_data_complete['school_name'].unique())
total_schools
total_students = len(school_data_complete['Student ID'].unique())
total_students
total_budget = school_data['budget'].sum()
total_budget

avg_math_score = student_data['math_score'].mean()
avg_math_score
avg_reading_score = student_data['reading_score'].mean()
avg_reading_score
overall_avg_score = (avg_math_score + avg_reading_score)/2
overall_avg_score

pct_passing_math = ((len(school_data_complete.loc[(school_data_complete['math_score'] >= 70)])) / total_students) * 100

pct_passing_reading = ((len(school_data_complete.loc[(school_data_complete['reading_score'] >= 70)])) / total_students) * 100

# Create dataframe

district_summary_table = pd.DataFrame({'Total Schools': [total_schools],
                                      'Total Students': [total_students],
                                      'Total Budget': [total_budget],
                                      'Average Math Score': [avg_math_score],
                                      'Average Reading Score': [avg_reading_score],
                                       '% Passing Math': [pct_passing_math],
                                       '% Passing Reading': [pct_passing_reading],
                                      'Overall Passing Rate': [overall_avg_score]})

# #Clean up numbers

district_summary_table['Total Students'] = district_summary_table['Total Students'].map("{:,}".format)
district_summary_table['Total Budget'] = district_summary_table['Total Budget'].map("${:,}".format)
district_summary_table['Average Math Score'] = district_summary_table['Average Math Score'].map("{:.2f}".format)
district_summary_table['Average Reading Score'] = district_summary_table['Average Reading Score'].map("{:.2f}".format)
district_summary_table['Overall Passing Rate'] = district_summary_table['Overall Passing Rate'].map("{:.2f}".format)
district_summary_table['% Passing Math'] = district_summary_table['% Passing Math'].map("{:.2f}".format)
district_summary_table['% Passing Reading'] = district_summary_table['% Passing Reading'].map("{:.2f}".format)

district_summary_table
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
# Set index and perform analysis functions 

school_types = school_data.set_index(['school_name'])['type']

school_type = school_data[['school_name', 'type']].groupby(['school_name'])['type']
school_type

students_per_school = school_data_complete['school_name'].value_counts()
students_per_school

budget_per_school = school_data_complete.groupby(['school_name']).mean()['budget']
budget_per_school
budget_per_student = budget_per_school / students_per_school


avg_math_score_by_school = school_data_complete.groupby(['school_name']).mean()['math_score']
avg_math_score_by_school
avg_reading_score_by_school = school_data_complete.groupby(['school_name']).mean()['reading_score']
avg_reading_score_by_school

passing_math_by_school = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name').count()['student_name']
passing_reading_by_school = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name').count()['student_name']
pct_passing_math_by_school = (passing_math_by_school / students_per_school) * 100
pct_passing_reading_by_school = (passing_reading_by_school / students_per_school) * 100
pct_passing_math_by_school
pct_passing_reading_by_school

overall_passing_rate_by_school = (pct_passing_math_by_school + pct_passing_reading_by_school) / 2
overall_passing_rate_by_school

# Create Dataframe

school_summary_df = pd.DataFrame({'School Type': school_type,
                                 'Total Students': students_per_school,
                                 'Total School Budget': budget_per_school,
                                 'Per Student Budget': budget_per_student,
                                 'Average Math Score': avg_math_score_by_school,
                                 'Average Reading Score': avg_reading_score_by_school,
                                 '% Passing Math': pct_passing_math_by_school,
                                 '% Passing Reading': pct_passing_reading_by_school,
                                 'Overall Passing Rate': overall_passing_rate_by_school})

school_summary_df
```

```python
# arrange by top performing schools, descending

top_performing_schools = school_summary_df.sort_values('Overall Passing Rate', ascending=False)
top_performing_schools.head(8)
```
